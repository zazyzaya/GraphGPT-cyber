```python
import matplotlib.pyplot as plt
```

```python
def plot_one(fname, name=''):
    with open(fname, 'r') as f:
        lines = f.read()
    lines = lines.split('\n')[:-1]
    tokens,losses = [],[]

    for line in lines:
        loss,step,token,time = line.split(',')
        losses.append(float(loss))
        tokens.append(float(token))

    plt.plot(tokens,losses, label=name)

HOME = 'pretrained/static_graph/rw_sampling'
plot_one(f'{HOME}/rw_bertlog_med.txt', name='Medium')
plot_one(f'{HOME}/rw_bertlog_mini.txt', name='Mini')

HOME = '.'
plot_one(f'{HOME}/rw_bertlog_mini.txt', name='Mini (temporal)')
plot_one(f'{HOME}/rw_bertlog_med.txt', name='Medium (temporal)')

plt.legend()
plt.show()
```
