```python
import matplotlib.pyplot as plt
```

```python
def plot_one(fname, name=''):
    with open(fname, 'r') as f:
        lines = f.read()
    lines = lines.split('\n')[:-1]
    tokens,losses = [],[]

    for line in lines:
        loss,step,token,time = line.split(',')
        losses.append(float(loss))
        tokens.append(float(token))

    plt.plot(tokens,losses, label=name)

plot_one('pretrained/bertlog_tiny.txt', name='Tiny')
plot_one('pretrained/bertlog_mini.txt', name='Mini')
plot_one('pretrained/bertlog_med.txt', name='Medium')
plt.legend()
plt.show()
```
