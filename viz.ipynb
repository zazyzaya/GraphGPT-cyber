```python
import matplotlib.pyplot as plt
```

```python
def plot_one(fname, name=''):
    with open(fname, 'r') as f:
        lines = f.read()
    lines = lines.split('\n')[:-1]
    tokens,losses = [],[]

    for line in lines:
        loss,step,token,time = line.split(',')
        losses.append(float(loss))
        tokens.append(float(token))

    plt.plot(tokens,losses, label=name)

HOME = '.'
plot_one(f'{HOME}/rw_bertlog_med.txt', name='RW')
plot_one(f'{HOME}/trw_bertlog_med.txt', name='TRW')


plt.legend()
plt.show()
```
